```python
import pandas as pd
import numpy as np
```

```python
df_mobil = pd.read_csv("https://raw.githubusercontent.com/xfrih/cars_pred/main/cars_data.csv")
df_mobil
```

```python
total_tanda_tanya = (df_mobil == "?").sum(axis=0)
total_tanda_tanya
```

```python
df_mobil = df_mobil.replace(to_replace="?", value=0)
df_mobil
```

```python
df_mobil.dtypes
```

```python
df_mobil[["horsepower", "price", "weight"]] = df_mobil[["horsepower", "price", "weight"]].astype(np.int64)
df_mobil.dtypes
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_mobil["highway-mpg"], df_mobil["price"])
plt.xlabel("highway-mpg")
plt.ylabel("price")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df_mobil[["highway-mpg", "weight", "horsepower"]]
y = df_mobil["price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model_regresi = LinearRegression()
model_regresi.fit(X_train, y_train)
```

```python
model_regresi_pred = model_regresi.predict(X_test)
```

```python
plt.scatter(X_test.iloc[:, 0], model_regresi_pred)
plt.xlabel("highway-mpg")
plt.ylabel("price")
plt.show
```
